Build GWR model

```python
import numpy as np
import pandas as pd
import libpysal as ps
from mgwr.gwr import GWR, MGWR
from mgwr.sel_bw import Sel_BW
from mgwr.utils import compare_surfaces, truncate_colormap
import geopandas as gp
import matplotlib.pyplot as plt
import matplotlib as mpl
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler
```

```python
# Chun b d liu u vo
y = df_train['AWS'].values.reshape((-1, 1))
X = df_train[['Imerg', 'R500', 'SLHF', 'TCLW', 'TCWV']].values

# Cc ta  khng gian
coords = df_train[['Row Position', 'Column Position']].values

# La chn bandwidth ti u
gwr_selector = Sel_BW(coords, y, X)
gwr_bw = gwr_selector.search(bw_min= 300, bw_max = 450)
print(gwr_bw)

# Calibrate m hnh GWR
model = GWR(coords, y, X, gwr_bw)
gwr_results = model.fit()
```

```python
# Chun b d liu khng gian v v cc h s
gdf = gp.GeoDataFrame(df_train, geometry=gp.points_from_xy(df_train['Column Position'], df_train['Row Position']))
gdf['Intercept'] = gwr_results.params[:, 0]
gdf['Imerg_coef'] = gwr_results.params[:, 1]
gdf['R500_coef'] = gwr_results.params[:, 2]
gdf['SLHF_coef'] = gwr_results.params[:, 3]
gdf['TCLW_coef'] = gwr_results.params[:, 4]
gdf['TCWV_coef'] = gwr_results.params[:, 5]

fig, axes = plt.subplots(3, 2, figsize=(15, 20))
columns = ['Intercept', 'Imerg_coef', 'R500_coef', 'SLHF_coef', 'TCLW_coef', 'TCWV_coef']  
for ax, column in zip(axes.flatten(), columns):
    gdf.plot(column=column, cmap='coolwarm', ax=ax, legend=True)
    ax.set_title(f'{column} spatial distribution')
    ax.invert_yaxis()  # o ngc trc y

plt.tight_layout()
plt.show()
```

```python
import pickle

# Lu model v trng s vo file
with open('/kaggle/working/gwr_model.pkl', 'wb') as f:
    pickle.dump(model, f)
    pickle.dump(gwr_results, f)
```

```python
# Kim tra v chun b d liu test cho df_test_1
y_test = df_test['AWS'].values.reshape((-1, 1))
X_test = df_test[['Imerg', 'R500', 'SLHF', 'TCLW', 'TCWV']]  
coords_test = df_test[['Row Position', 'Column Position']].values

# D on gi tr cho tp test
scale = gwr_results.scale
residuals = gwr_results.resid_response
pred_y = model.predict(coords_test, X_test, scale, residuals)
pred_y.predictions[pred_y.predictions < 0] = 0
```

```python
# Hin th kt qu
plt.scatter(y_test, pred_y.predictions)
plt.xlabel('Actual AWS')
plt.ylabel('Predicted AWS')
plt.title('Actual vs. Predicted AWS for df_test')
plt.show()
```

```python
# Lu kt qu d on vo tp CSV
results_df = pd.DataFrame({
    'Actual AWS': y_test.flatten(),
    'Predicted AWS': pred_y.predictions.flatten()
})
results_df.to_csv('predictions_GWR.csv', index=False)
```
