```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load the dataset from your folder
# Update the path to your actual file location
file_path = '/content/data_test.csv'
data_test = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
print(data_test.head())

file_path = '/content/data_train.csv'
data_train = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
print(data_train.head())


X_train = data_train.drop(['AWS', 'Row Position', 'Column Position', 'DEM'], axis=1)
y_train = data_train['AWS']
X_test = data_test.drop(['AWS', 'Row Position', 'Column Position', 'DEM'], axis=1)
y_test = data_test['AWS']
name_test = data_train['name']
AWS = data_test['AWS']

# Initialize the Random Forest Regressor with out-of-bag samples
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42, oob_score=True)

# Train the model
rf_regressor.fit(X_train, y_train)

# Make predictions
y_pred = rf_regressor.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')

# Compare actual vs predicted values
comparison_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

# Visualize the comparison
plt.figure(figsize=(10, 6))
plt.plot(comparison_df['Actual'].values, label='Actual', color='b', alpha=0.6)
plt.plot(comparison_df['Predicted'].values, label='Predicted', color='r', alpha=0.6)
plt.xlabel('Sample Index')
plt.ylabel('Rainfall')
plt.title('Actual vs Predicted Rainfall')
plt.legend()
plt.show()

# Scatter plot for better visualization
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--')
plt.xlabel('Actual Rainfall')
plt.ylabel('Predicted Rainfall')
plt.title('Actual vs Predicted Rainfall (Scatter Plot)')
plt.show()

# Feature importance ranking
importances = rf_regressor.feature_importances_
feature_names = X_train.columns
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

print('Feature Importances:')
print(importance_df)

# Iteratively remove less important features
threshold = 0.01  # Set a threshold for feature importance
while True:
    # Remove features with importance below the threshold
    features_to_keep = importance_df[importance_df['Importance'] >= threshold]['Feature']

    if len(features_to_keep) == len(feature_names):
        break  # Stop if no features are removed

    feature_names = features_to_keep
    features = data_train[feature_names]

    # Split the data again
    X_train, X_test, y_train, y_test = train_test_split(features, y_train, test_size=0.4, random_state=42)

    # Train a new model
    rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42, oob_score=True)
    rf_regressor.fit(X_train, y_train)

    # Update feature importances
    importances = rf_regressor.feature_importances_
    importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
    importance_df = importance_df.sort_values(by='Importance', ascending=False)

    print('Updated Feature Importances:')
    print(importance_df)

# Final model training with optimized feature set
final_features = importance_df['Feature']
X_train, X_test, y_train, y_test = train_test_split(data_train[final_features], y_train, test_size=0.4, random_state=42)

final_rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42, oob_score=True)
final_rf_regressor.fit(X_train, y_train)

# Final predictions
final_y_pred = final_rf_regressor.predict(X_test)

# Final evaluation
final_mse = mean_squared_error(y_test, final_y_pred)
final_r2 = r2_score(y_test, final_y_pred)
print(f'Final Mean Squared Error: {final_mse:.2f}')
print(f'Final R-squared: {final_r2:.2f}')

# Show final evaluation table
final_comparison_df = pd.DataFrame({
    'AWS (Actual)': AWS,
    'Predicted': final_y_pred
})
# final_comparison_df = final_comparison_df.sort_values(by='Predicted', ascending=False)

print(f'{final_comparison_df}')

final_comparison_df.to_csv('RandomForest.csv', index=False)

# Step 5: Download the CSV file
from google.colab import files
files.download('RandomForest.csv')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr


def load_and_clean_data(filename):
    data = pd.read_csv(filename)
    # Ensure the data columns are numeric
    data = data.apply(pd.to_numeric, errors='coerce')
    # Drop rows with any NaN values
    data = data.dropna()
    return data

# Load and clean data from CSV files

# Load data from CSV files
file_path_rf = '/content/predictions_RF.csv'
data_rf = load_and_clean_data(file_path_rf)
rf = data_rf['Predicted']

file_path_gwr = '/content/predictions_GWR.csv'
data_gwr = load_and_clean_data(file_path_gwr)
gwr = data_gwr['Predicted AWS']

file_path_step = '/content/predictions_step.csv'
data_step = load_and_clean_data(file_path_step)
step = data_step['y_pred']

file_path = '/content/data_test.csv'
data = load_and_clean_data(file_path)
imerg = data['Imerg']
obs = data['AWS']

# Function to calculate Taylor diagram statistics
def calculate_statistics(observed, model):
    std_dev = np.std(model)
    corr_coeff = pearsonr(observed, model)[0]
    rmse = np.sqrt(mean_squared_error(observed, model))
    return std_dev, corr_coeff, rmse

# Calculate statistics for each model
std_obs = np.std(obs)
stats = {}
for model_name, model_data in zip(['IMERG', 'STEP', 'GWR', 'RF'], [imerg, step, gwr, rf]):
    std_dev, corr_coeff, _ = calculate_statistics(obs, model_data)
    stats[model_name] = (std_dev, corr_coeff)

# Extract the statistics
model_labels = list(stats.keys())
std_dev = np.array([stats[model][0] for model in model_labels])
corr_coeff = np.array([stats[model][1] for model in model_labels])

# Function to plot Taylor diagram
def taylor_diagram(std_dev, corr_coeff, std_obs, model_labels):
    """
    Plot Taylor diagram.

    Args:
        std_dev (array-like): Standard deviations of model predictions.
        corr_coeff (array-like): Correlation coefficients of model predictions.
        std_obs (float): Standard deviation of observed data.
        model_labels (list): Labels for each model.
    """
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, polar=True)

    # Define the angular coordinates for correlation coefficient
    theta = np.arccos(corr_coeff)

    # Plot the points
    for i in range(len(std_dev)):
        ax.plot(theta[i], std_dev[i] / std_obs, 'o', markersize=10, label=model_labels[i])

    # Plot reference line for std_obs
    ax.plot([0, np.pi/2], [1, 1], '--', color='grey')

    # Set the radial ticks (normalized standard deviation)
    ax.set_rticks([0.5, 1, 1.5, 2])
    ax.set_yticklabels(['0.5', '1.0', '1.5', '2.0'])

    # Set the angular ticks (correlation coefficient)
    ax.set_xticks(np.arccos([0.1, 0.4, 0.7, 0.9]))
    ax.set_xticklabels(['0.1', '0.4', '0.7', '0.9'])

    # Add labels and legend
    ax.set_ylabel('Normalized Standard Deviation')
    plt.legend(loc='upper right')

    # Add title
    plt.title('Taylor Diagram')

    # Show the plot
    plt.show()

# Plot the Taylor diagram
taylor_diagram(std_dev, corr_coeff, std_obs, model_labels)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr

# Load data from CSV file

file_path_rf = '/content/data_rf2_file.csv'
data_rf = pd.read_csv(file_path_rf)
rf = data_rf['Predicted']

file_path_gwr = '/content/data_gwr2_file.csv'
data_gwr = pd.read_csv(file_path_gwr)
gwr = data_gwr['Predicted AWS']

file_path_step = '/content/data_step2_file.csv'
data_step = pd.read_csv(file_path_step)
step = data_step['y_pred']

file_path = '/content/data_test2_file.csv'
data_test = pd.read_csv(file_path)
imerg = data_test['Imerg']
obs = data_test['AWS']
date = data_test['name']

# Function to calculate NSE
def nash_sutcliffe_efficiency(observed, model):
    return 1 - (np.sum((observed - model) ** 2) / np.sum((observed - np.mean(observed)) ** 2))

# Function to calculate metrics for each model
def calculate_metrics(observed, model):
    cc = pearsonr(observed, model)[0]
    nse = nash_sutcliffe_efficiency(observed, model)
    rmse = np.sqrt(mean_squared_error(observed, model))
    return cc, nse, rmse

# Calculate metrics for each model
metrics = {
    'date': date,
    'CC_IMERG': [],
    'NSE_IMERG': [],
    'RMSE_IMERG': [],
    'CC_STEP': [],
    'NSE_STEP': [],
    'RMSE_STEP': [],
    'CC_GWR': [],
    'NSE_GWR': [],
    'RMSE_GWR': [],
    'CC_RF': [],
    'NSE_RF': [],
    'RMSE_RF': []
}

for i in range(len(date)):
    if i == 0:
        metrics['CC_IMERG'].append(np.nan)
        metrics['NSE_IMERG'].append(np.nan)
        metrics['RMSE_IMERG'].append(np.nan)
        metrics['CC_STEP'].append(np.nan)
        metrics['NSE_STEP'].append(np.nan)
        metrics['RMSE_STEP'].append(np.nan)
        metrics['CC_GWR'].append(np.nan)
        metrics['NSE_GWR'].append(np.nan)
        metrics['RMSE_GWR'].append(np.nan)
        metrics['CC_RF'].append(np.nan)
        metrics['NSE_RF'].append(np.nan)
        metrics['RMSE_RF'].append(np.nan)
    else:
        metrics['CC_IMERG'].append(calculate_metrics(obs[:i+1], imerg[:i+1])[0])
        metrics['NSE_IMERG'].append(calculate_metrics(obs[:i+1], imerg[:i+1])[1])
        metrics['RMSE_IMERG'].append(calculate_metrics(obs[:i+1], imerg[:i+1])[2])
        metrics['CC_STEP'].append(calculate_metrics(obs[:i+1], step[:i+1])[0])
        metrics['NSE_STEP'].append(calculate_metrics(obs[:i+1], step[:i+1])[1])
        metrics['RMSE_STEP'].append(calculate_metrics(obs[:i+1], step[:i+1])[2])
        metrics['CC_GWR'].append(calculate_metrics(obs[:i+1], gwr[:i+1])[0])
        metrics['NSE_GWR'].append(calculate_metrics(obs[:i+1], gwr[:i+1])[1])
        metrics['RMSE_GWR'].append(calculate_metrics(obs[:i+1], gwr[:i+1])[2])
        metrics['CC_RF'].append(calculate_metrics(obs[:i+1], rf[:i+1])[0])
        metrics['NSE_RF'].append(calculate_metrics(obs[:i+1], rf[:i+1])[1])
        metrics['RMSE_RF'].append(calculate_metrics(obs[:i+1], rf[:i+1])[2])

metrics_df = pd.DataFrame(metrics)

# Plot the metrics
plt.figure(figsize=(12, 18))

# Plot Correlation Coefficient (CC)
plt.subplot(3, 1, 1)
plt.plot(metrics_df['date'], metrics_df['CC_IMERG'], label='IMERG', marker='^', linestyle='-')
plt.plot(metrics_df['date'], metrics_df['CC_STEP'], label='STEP', marker='^', linestyle='-')
plt.plot(metrics_df['date'], metrics_df['CC_GWR'], label='GWR', marker='^', linestyle='-')
plt.plot(metrics_df['date'], metrics_df['CC_RF'], label='RF', marker='^', linestyle='-')
plt.title('Correlation Coefficient (CC)')
plt.xlabel('Date')
plt.ylabel('CC')
plt.legend()
plt.grid()

# Plot Nash-Sutcliffe Efficiency (NSE)
plt.subplot(3, 1, 2)
plt.plot(metrics_df['date'], metrics_df['NSE_IMERG'], label='IMERG', marker='^', linestyle='-')
plt.plot(metrics_df['date'], metrics_df['NSE_STEP'], label='STEP', marker='^', linestyle='-')
plt.plot(metrics_df['date'], metrics_df['NSE_GWR'], label='GWR', marker='^', linestyle='-')
plt.plot(metrics_df['date'], metrics_df['NSE_RF'], label='RF', marker='^', linestyle='-')
plt.title('Nash-Sutcliffe Efficiency (NSE)')
plt.xlabel('Date')
plt.ylabel('NSE')
plt.legend()
plt.grid()

# Plot Root Mean Squared Error (RMSE)
plt.subplot(3, 1, 3)
plt.plot(metrics_df['date'], metrics_df['RMSE_IMERG'], label='IMERG', marker='^', linestyle='-')
plt.plot(metrics_df['date'], metrics_df['RMSE_STEP'], label='STEP', marker='^', linestyle='-')
plt.plot(metrics_df['date'], metrics_df['RMSE_GWR'], label='GWR', marker='^', linestyle='-')
plt.plot(metrics_df['date'], metrics_df['RMSE_RF'], label='RF', marker='^', linestyle='-')
plt.title('Root Mean Squared Error (RMSE)')
plt.xlabel('Date')
plt.ylabel('RMSE (mm)')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
